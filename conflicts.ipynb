```python
def group_directly_intersecting_sets(sets_list):
    """
    Group sets that directly intersect. Sets are only grouped if they share 
    at least one element directly (no transitive grouping).
    
    Example:
        Input: [{1,2}, {2,3}, {3,4}, {5,6}]
        Output: [[{1,2}, {2,3}], [{2,3}, {3,4}], [{5,6}]]
        Note: {1,2} and {3,4} are NOT grouped together even though 
              they're connected through {2,3}
    """
    groups = []
    
    for s in sets_list:
        # Find groups where at least one set directly intersects with current set
        matching_groups = []
        remaining_groups = []
        
        for group in groups:
            # Check if current set directly intersects with ANY set in this group
            if any(s & group_set for group_set in group):
                matching_groups.append(group)
            else:
                remaining_groups.append(group)
        
        # Add current set to all matching groups (or create new group)
        if matching_groups:
            # Add to all groups that have direct intersection
            for group in matching_groups:
                group.append(s)
        else:
            # No direct intersection, create new group
            groups.append([s])
    
    return groups


# Example usage
example_sets = [
    {1, 2, 3},      # A
    {3, 4, 5},      # B - intersects A (shares 3)
    {5, 6},         # C - intersects B (shares 5), but NOT A
    {7, 8},         # D
    {8, 9},         # E - intersects D (shares 8)
    {10}            # F - no intersections
]

grouped = group_directly_intersecting_sets(example_sets)
for i, group in enumerate(grouped):
    print(f"Group {i}: {group}")
```
